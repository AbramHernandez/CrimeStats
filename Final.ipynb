```python
import pandas as pd
import plotnine as pn

file_path = r"C:\Users\Ilike\CSCI385\Final Project\Hate Crime 2012 - 2022.csv"
data = pd.read_csv(file_path)

race_ethnicity_ancestry_data = data[data['MostSeriousBiasType'] == 'Race/Ethnicity/Ancestry']

# Create a list of years from 2001 to 2023
years = list(range(2001, 2024)) 

(pn.ggplot(race_ethnicity_ancestry_data)
+ pn.aes(x='ClosedYear')
+ pn.geom_histogram(bins=len(years))  
+ pn.labs(title='Distribution of Hate Crimes Based on Race/Ethnicity/Ancestry Over Time (2001-2023)', x='Year', y='Number of Hate Crimes')
+ pn.theme_bw()
+ pn.scale_x_continuous(breaks=years)  
)
```
